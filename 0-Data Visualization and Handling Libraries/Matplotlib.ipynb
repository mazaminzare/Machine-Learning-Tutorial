````python
import matplotlib.pyplot as plt
import numpy as np
````

## 2. Basic Plotting

````python
# Creating a simple line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
````

## 3. Subplots

````python
# Creating multiple plots in one figure
fig, axes = plt.subplots(2, 1)
axes[0].plot(x, y, 'r')
axes[1].plot(x, np.cos(x), 'b')

axes[0].set_title('Sine Wave')
axes[1].set_title('Cosine Wave')

plt.tight_layout()
plt.show()
````

## 4. Bar Charts

````python
# Creating a bar chart
categories = ['Apples', 'Bananas', 'Cherries']
values = [20, 33, 15]

plt.figure()
plt.bar(categories, values)
plt.title('Fruit Count')
plt.xlabel('Fruit')
plt.ylabel('Quantity')
plt.show()
````

## 5. Histograms

````python
# Creating a histogram
data = np.random.normal(0, 1, size=1000)

plt.figure()
plt.hist(data, bins=30)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

## 6. Scatter Plots

````python
# Creating a scatter plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure()
plt.scatter(x, y)
plt.title('Random Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
````

## 7. Pie Charts

````python
# Creating a pie chart
sizes = [215, 130, 245, 210]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']

plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart of Fruits')
plt.show()
````

## 8. Customization and Styles

````python
# Customizing plots with styles
plt.style.use('ggplot')

x = np.linspace(0, 10, 100)
y = np.sin(x) * np.cos(x)

plt.figure()
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.title('Custom Style Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
````

## 9. Advanced Plotting Techniques

````python
# Advanced plotting with error bars
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.figure()
plt.errorbar(x, y, yerr=dy, fmt='.k')
plt.title('Plot with Error Bars')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
````
