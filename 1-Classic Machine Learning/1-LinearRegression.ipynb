## Linear Regression Formula

Linear regression is used to model the relationship between a dependent variable $y$ and one or more explanatory variables (denoted $x$). The formula for simple linear regression (one dependent and one independent variable) is:

$$
y = \beta_0 + \beta_1 x + \epsilon
$$

where:
- $y$ is the dependent variable,
- $x$ is the independent variable,
- $\beta_0$ is the intercept,
- $\beta_1$ is the slope,
- $\epsilon$ is the error term.

### Least Squares Estimation

The parameters $\beta_0$ and $\beta_1$ are estimated using the least squares method, which minimizes the sum of squared residuals. The equations are:

$$
\beta_1 = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^n (x_i - \bar{x})^2}
$$

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

where $\bar{x}$ and $\bar{y}$ are the sample means of $x$ and $y$, respectively.

### Numerical Example

Consider the dataset:

| x   | y   |
|-----|-----|
| 1   | 2   |
| 2   | 4   |
| 3   | 6   |

**Step 1:** Calculate the means $\bar{x}$ and $\bar{y}$:

$$
\bar{x} = \frac{1+2+3}{3} = 2
$$

$$
\bar{y} = \frac{2+4+6}{3} = 4
$$

**Step 2:** Calculate $\beta_1$:

$$
\beta_1 = \frac{(1-2)(2-4) + (2-2)(4-4) + (3-2)(6-4)}{(1-2)^2 + (2-2)^2 + (3-2)^2} = \frac{(-1)(-2) + 0 + 1 \times 2}{1 + 0 + 1} = 2
$$

**Step 3:** Calculate $\beta_0$:

$$
\beta_0 = 4 - 2 \times 2 = 0
$$

Thus, the linear regression model based on this dataset is:

$$
y = 0 + 2x
$$

This example demonstrates a perfect linear relationship between $x$ and $y$ in the given dataset.


## Linear Regression in Machine Learning

Linear regression in machine learning is typically approached as an optimization problem where the objective is to minimize the Mean Squared Error (MSE) between the predicted and actual values. Here's an in-depth explanation, including both gradient descent and the closed-form solution.

### Mean Squared Error (MSE)

The loss function used in linear regression is the MSE, which quantifies the difference between the observed values and the values predicted by the model:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$

where $y_i$ are the actual values and $\hat{y}_i = \beta_0 + \beta_1 x_i$ are the predicted values.

### Gradient Descent

Gradient descent is used to minimize the MSE. The parameters $\beta_0$ and $\beta_1$ are updated iteratively based on the gradients of the MSE.

#### Gradient Calculation

The gradients of the MSE with respect to $\beta_0$ and $\beta_1$ are:

$$
\frac{\partial \text{MSE}}{\partial \beta_0} = -\frac{2}{n} \sum_{i=1}^n (y_i - (\beta_0 + \beta_1 x_i))
$$

$$
\frac{\partial \text{MSE}}{\partial \beta_1} = -\frac{2}{n} \sum_{i=1}^n (y_i - (\beta_0 + \beta_1 x_i)) x_i
$$

#### Update Rules

Using these gradients, the parameters are updated as follows:

$$
\beta_0 := \beta_0 - \alpha \left(-\frac{2}{n} \sum_{i=1}^n (y_i - (\beta_0 + \beta_1 x_i))\right)
$$

$$
\beta_1 := \beta_1 - \alpha \left(-\frac{2}{n} \sum_{i=1}^n (y_i - (\beta_0 + \beta_1 x_i)) x_i\right)
$$

where $\alpha$ is the learning rate.

### Closed-Form Solution

The closed-form solution, also known as the normal equation, provides a direct way to find the optimal parameters:

$$
\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
$$

Here, $\mathbf{X}$ is the matrix of input features (including a column of ones for the intercept), and $\mathbf{y}$ is the vector of output values. This method requires computing the inverse of $\mathbf{X}^T \mathbf{X}$, which can be computationally expensive for large datasets.

### Example with Gradient Descent

1. **Initialize Parameters**: Start with random values for $\beta_0$ and $\beta_1$.
2. **Calculate Error and Gradient**: Compute the MSE and its gradients with respect to $\beta_0$ and $\beta_1$.
3. **Update Parameters**: Adjust $\beta_0$ and $\beta_1$ using the update rules.
4. **Repeat**: Repeat steps 2 and 3 until the changes in MSE are minimal or a certain number of iterations is reached.

This machine learning approach iteratively adjusts the coefficients to minimize the MSE, providing an efficient and scalable method for fitting the linear regression model.


### Numerical Example for Gradient Descent

Assume we have the following dataset:

| x   | y   |
|-----|-----|
| 1   | 3   |
| 2   | 5   |
| 3   | 7   |

We will use gradient descent to find the best fitting line $y = \beta_0 + \beta_1 x$.

#### Step 1: Initialize Parameters

Let's start with initial guesses for the parameters:
- $\beta_0 = 0$
- $\beta_1 = 0$

#### Step 2: Set Learning Rate

Choose a learning rate, $\alpha = 0.01$. This value is small to ensure that the gradient descent does not overshoot the minimum.

#### Step 3: Perform Gradient Descent

We'll perform a single iteration of gradient descent to update $\beta_0$ and $\beta_1$. Normally, you would repeat this process until convergence, but for simplicity, we'll show just one iteration.

##### Calculate the Gradients

The gradients are calculated as follows:

$$
\frac{\partial \text{MSE}}{\partial \beta_0} = -\frac{2}{n} \sum_{i=1}^n (y_i - (\beta_0 + \beta_1 x_i))
$$

$$
\frac{\partial \text{MSE}}{\partial \beta_1} = -\frac{2}{n} \sum_{i=1}^n (y_i - (\beta_0 + \beta_1 x_i)) x_i
$$

Plugging in the values:

- $n = 3$
- $\beta_0 = 0$
- $\beta_1 = 0$

The gradients become:

$$
\frac{\partial \text{MSE}}{\partial \beta_0} = -\frac{2}{3} [(3 - 0) + (5 - 0) + (7 - 0)] = -\frac{2}{3} \times 15 = -10
$$

$$
\frac{\partial \text{MSE}}{\partial \beta_1} = -\frac{2}{3} [(3 - 0) \times 1 + (5 - 0) \times 2 + (7 - 0) \times 3] = -\frac{2}{3} \times 34 = -22.67
$$

##### Update Parameters

Using the learning rate $\alpha = 0.01$:

$$
\beta_0 := \beta_0 - \alpha \times (-10) = 0 + 0.1 = 0.1
$$

$$
\beta_1 := \beta_1 - \alpha \times (-22.67) = 0 + 0.2267 = 0.2267
$$

#### Conclusion

After just one iteration, the updated parameters are:
- $\beta_0 = 0.1$
- $\beta_1 = 0.2267$

To achieve convergence to the optimal parameters, continue iterating until the changes in $\beta_0$ and $\beta_1$ become minimal.

This numerical example demonstrates the gradient descent method in action for fitting a linear regression model, iteratively adjusting the coefficients to minimize the MSE.


### Closed-Form Solution for Linear Regression

The closed-form solution, also known as the normal equation, is a non-iterative method for finding the optimal parameters of a linear regression model. This method provides a direct calculation of the parameters that minimize the cost function.

#### Formula for the Normal Equation

The formula to compute the regression coefficients in the closed-form is:

$$
\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
$$

Here:
- $\mathbf{X}$ is the matrix of input features, augmented with a column of ones for the intercept.
- $\mathbf{y}$ is the vector of output values.

This method involves matrix operations and is best used when the number of features is not too large, as the computation of $(\mathbf{X}^T \mathbf{X})^{-1}$ can be computationally expensive and numerically unstable.

#### Numerical Example

Consider the same dataset used previously:

| x   | y   |
|-----|-----|
| 1   | 3   |
| 2   | 5   |
| 3   | 7   |

To apply the normal equation, we first need to construct the matrix $\mathbf{X}$ (with an additional column of ones for the intercept) and the vector $\mathbf{y}$.

##### Construct $\mathbf{X}$ and $\mathbf{y}$

$$
\mathbf{X} = \begin{bmatrix}
1 & 1 \\
1 & 2 \\
1 & 3 \\
\end{bmatrix},
\quad
\mathbf{y} = \begin{bmatrix}
3 \\
5 \\
7 \\
\end{bmatrix}
$$

##### Compute $\boldsymbol{\beta}$

$$
\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
$$

Calculating $\mathbf{X}^T \mathbf{X}$ and its inverse:

$$
\mathbf{X}^T \mathbf{X} = \begin{bmatrix}
3 & 6 \\
6 & 14 \\
\end{bmatrix}
$$

$$
(\mathbf{X}^T \mathbf{X})^{-1} = \begin{bmatrix}
1.4 & -0.6 \\
-0.6 & 0.3 \\
\end{bmatrix}
$$

Now, compute $\mathbf{X}^T \mathbf{y}$:

$$
\mathbf{X}^T \mathbf{y} = \begin{bmatrix}
15 \\
35 \\
\end{bmatrix}
$$

Finally, calculate $\boldsymbol{\beta}$:

$$
\boldsymbol{\beta} = \begin{bmatrix}
1.4 & -0.6 \\
-0.6 & 0.3 \\
\end{bmatrix} \times \begin{bmatrix}
15 \\
35 \\
\end{bmatrix} = \begin{bmatrix}
1 \\
2 \\
\end{bmatrix}
$$

#### Conclusion

The calculated coefficients are $\beta_0 = 1$ and $\beta_1 = 2$. The regression line is thus $y = 1 + 2x$. This demonstrates that the closed-form solution directly computes the optimal parameters for linear regression, avoiding the iterative process required by methods like gradient descent.

This example provides a complete implementation of the normal equation in linear regression, offering a direct method for obtaining the regression coefficients.


## Assumptions Behind Linear Regression

Linear regression relies on several key assumptions:

1. **Linearity**: The relationship between the independent and dependent variables is assumed to be linear.
2. **Independence**: Observations are assumed to be independent of each other.
3. **Homoscedasticity**: The variance of residual errors is constant across all levels of the independent variables.
4. **Normal Distribution of Errors**: Residuals are normally distributed, particularly for the purpose of hypothesis testing.
5. **No Perfect Multicollinearity**: The independent variables should not be perfectly correlated.

## Bias-Variance Trade-off

Linear regression typically exhibits:

- **High Bias**: Due to its linear nature, if the true relationship in the data is not linear, the model will not capture complex patterns, leading to underfitting.
- **Low Variance**: It does not learn excessively from the noise in the training data, thus showing similar performance on new unseen data.

The simplicity of linear regression often results in a model that has high bias but low variance, making it prone to underfitting but stable across different samples of data.

## Advantages of Linear Regression

1. **Simplicity**: It is straightforward to understand and interpret.
2. **Efficiency**: Being computationally efficient, especially with the closed-form solution.
3. **Low Overfitting Risk**: Generally less prone to overfitting, particularly in scenarios with fewer predictor variables.
4. **Interpretability**: Provides clear insights into the relationship between variables.

## Disadvantages of Linear Regression

1. **Outlier Sensitivity**: Linear regression is sensitive to outliers, which can significantly influence the model fit.
2. **Limited to Linear Relationships**: It can only model linear relationships unless transformations are applied to the data.
3. **Multicollinearity Issues**: High correlations among predictor variables can lead to unreliable and unstable estimates of regression coefficients.
4. **Homoscedasticity Requirement**: Non-constant variance of residuals can lead to inefficiencies and biases in the coefficient estimates.

This comprehensive view helps in understanding when linear regression is appropriate and highlights the need for careful data preparation and analysis to ensure robust results.


```python
import numpy as np
import matplotlib.pyplot as plt

# Create synthetic data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # 100 points
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3x + noise

# Plot the data
plt.scatter(X, y, label='Data Points')
plt.title("Synthetic Linear Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Adding x0 = 1 to each instance
X_b = np.c_[np.ones((100, 1)), X]  # Add x0 = 1 to each instance

# Normal Equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print("Theta from Normal Equation:", theta_best.flatten())
```

```python
# Make predictions using the Normal Equation results
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # Add intercept term
y_predict = X_new_b.dot(theta_best)

# Plot the data and the model prediction
plt.scatter(X, y, alpha=0.5, label='Data Points')
plt.plot(X_new, y_predict, 'r-', linewidth=2, label='Prediction (Normal Equation)')
plt.title('Linear Regression Fit Using Closed-Form Solution')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Output the parameters
print("Calculated parameters (theta):")
print(theta_best)
```

```python
# Gradient Descent settings
eta = 0.1  # learning rate
n_iterations = 1000
m = 100  # number of instances

theta = np.random.randn(2, 1)  # random initialization

for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta -= eta * gradients

print("Theta from Gradient Descent:", theta.flatten())
```

```python
# Prepare data for plotting
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict_normal = X_new_b.dot(theta_best)
y_predict_gradient = X_new_b.dot(theta)

# Plotting the predictions
plt.scatter(X, y, alpha=0.5, label='Data Points')
plt.plot(X_new, y_predict_normal, "r-", linewidth=2, label="Normal Equation")
plt.plot(X_new, y_predict_gradient, "b--", linewidth=2, label="Gradient Descent")
plt.title("Linear Regression Comparisons")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
