```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulated data: 100 coin flips with 61 heads (1) and 39 tails (0)
data = np.concatenate([np.ones(61), np.zeros(39)])
n_heads = np.sum(data)
n_tails = len(data) - n_heads

# Define prior parameters for Beta distribution
alpha_prior = 1
beta_prior = 1

# Update parameters with observed data
alpha_post = alpha_prior + n_heads
beta_post = beta_prior + n_tails

# Posterior distribution
x = np.linspace(0, 1, 100)
posterior = stats.beta.pdf(x, alpha_post, beta_post)

# Plot the prior and posterior distributions
plt.figure(figsize=(10, 5))
plt.plot(x, stats.beta.pdf(x, alpha_prior, beta_prior), label='Prior (Beta(1, 1))')
plt.plot(x, posterior, label=f'Posterior (Beta({alpha_post}, {beta_post}))')
plt.xlabel('p')
plt.ylabel('Density')
plt.legend()
plt.title('Prior and Posterior Distributions')
plt.show()
```
